```python
import requests 
import pandas as pd
import numpy as np
from lxml import etree
```

```python
url = 'https://www.jianshu.com/recommendations/users?utm_source=desktop&utm_medium=index-users'


headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36'}
res = requests.get(url,headers = headers)
res.text


parsed = etree.HTML(res.text)
users = parsed.xpath('/html/body/div[1]/div/div/div/a[1]/h4')
users = [u.text.split()[0] for u in users]
users

hrefs = parsed.xpath('/html/body/div[1]/div/div/div/a[1]/@href')
hrefs = [h[7:]    for h in hrefs ]
hrefs
```

```python
data_users = pd.DataFrame()
for h in hrefs:
    base_url = 'https://www.jianshu.com/u/'
    new_url = base_url + h
    res = requests.get(new_url,headers = headers)
    parsed = etree.HTML(res.text)
    first3num = parsed.xpath('//ul/li/div/a/p/text()')
    print(first3num)
    last3num = parsed.xpath('//div[@class="info"]/ul/li/div/p/text()')
    print(last3num)
    
    data = pd.DataFrame()
    data['关注'] = [first3num[0]]
    data['粉丝'] = [first3num[1]]
    data['文章'] = [first3num[2]]
    
    data['字数'] = [last3num[0]]
    data['收获喜欢'] = [last3num[1]]
    data['总资产'] = [last3num[2]]
    data_users = pd.concat([data_users,data],ignore_index=True)
data_users
```
